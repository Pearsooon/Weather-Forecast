```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
```

```python
PROJECT_ROOT = os.path.abspath("..")
sys.path.append(PROJECT_ROOT)

from scripts.utils import execute_query

# Set plot style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("‚úÖ Libraries imported successfully")
```

```python
query = """
SELECT * 
FROM RAW.WEATHER_RAW 
LIMIT 100000
"""

df = execute_query(query, config_path='../config/snowflake_config.json')
print(f"‚úÖ Loaded {len(df):,} records from Snowflake")
print(f"\nüìä Dataset Shape: {df.shape}")
```

```python
# Basic information about the dataset
print("=" * 80)
print("DATA OVERVIEW")
print("=" * 80)

print("\n1. Column Data Types:")
print(df.dtypes)

print("\n2. Memory Usage:")
print(df.memory_usage(deep=True))

print("\n3. Dataset Info:")
df.info()
```

```python
# Data overview
print("=" * 80)
print("SAMPLE DATA")
print("=" * 80)

print("\nFirst 10 rows:")
display(df.head(10))

print("\nLast 10 rows:")
display(df.tail(10))

print("\nRandom 10 rows:")
display(df.sample(10))
```

```python
#Statistical Summary
print("=" * 80)
print("STATISTICAL SUMMARY")
print("=" * 80)

print("\nNumerical Columns:")
display(df.describe())

print("\nCategorical Columns:")
display(df.describe(include=['object']))
```

```python
#Data Distribution by Location
print("=" * 80)
print("DATA DISTRIBUTION BY LOCATION")
print("=" * 80)

location_stats = df.groupby('location_name').agg({
    'temperature': ['count', 'mean', 'std', 'min', 'max'],
    'precipitation': ['sum', 'mean'],
    'humidity': 'mean',
    'wind_speed': 'mean'
}).round(2)

display(location_stats)

# Visualize
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Records per location
df['location_name'].value_counts().plot(kind='bar', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Number of Records per Location', fontweight='bold')
axes[0,0].set_ylabel('Count')
axes[0,0].tick_params(axis='x', rotation=45)

# Average temperature by location
df.groupby('location_name')['temperature'].mean().plot(kind='bar', ax=axes[0,1], color='coral')
axes[0,1].set_title('Average Temperature by Location', fontweight='bold')
axes[0,1].set_ylabel('Temperature (¬∞C)')
axes[0,1].tick_params(axis='x', rotation=45)

# Total precipitation by location
df.groupby('location_name')['precipitation'].sum().plot(kind='bar', ax=axes[1,0], color='lightgreen')
axes[1,0].set_title('Total Precipitation by Location', fontweight='bold')
axes[1,0].set_ylabel('Precipitation (mm)')
axes[1,0].tick_params(axis='x', rotation=45)

# Average humidity by location
df.groupby('location_name')['humidity'].mean().plot(kind='bar', ax=axes[1,1], color='lightblue')
axes[1,1].set_title('Average Humidity by Location', fontweight='bold')
axes[1,1].set_ylabel('Humidity (%)')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Temporal Analysis
print("=" * 80)
print("TEMPORAL ANALYSIS")
print("=" * 80)

# Convert datetime
df['datetime'] = pd.to_datetime(df['datetime'])
df['date'] = df['datetime'].dt.date
df['hour'] = df['datetime'].dt.hour
df['day_of_week'] = df['datetime'].dt.dayofweek
df['month'] = df['datetime'].dt.month
df['year'] = df['datetime'].dt.year

print(f"\nDate Range: {df['datetime'].min()} to {df['datetime'].max()}")
print(f"Total Days: {(df['datetime'].max() - df['datetime'].min()).days}")
print(f"Total Hours: {len(df['datetime'].unique())}")

# Time distribution
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Records by hour
df['hour'].value_counts().sort_index().plot(kind='bar', ax=axes[0,0], color='orange')
axes[0,0].set_title('Records by Hour of Day', fontweight='bold')
axes[0,0].set_xlabel('Hour')
axes[0,0].set_ylabel('Count')

# Records by day of week
day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
df['day_of_week'].value_counts().sort_index().plot(kind='bar', ax=axes[0,1], color='purple')
axes[0,1].set_title('Records by Day of Week', fontweight='bold')
axes[0,1].set_xticklabels(day_names, rotation=0)
axes[0,1].set_ylabel('Count')

# Records by month
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['month'].value_counts().sort_index().plot(kind='bar', ax=axes[1,0], color='teal')
axes[1,0].set_title('Records by Month', fontweight='bold')
axes[1,0].set_xticklabels(month_names, rotation=45)
axes[1,0].set_ylabel('Count')

# Records by year
df['year'].value_counts().sort_index().plot(kind='bar', ax=axes[1,1], color='brown')
axes[1,1].set_title('Records by Year', fontweight='bold')
axes[1,1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Variable Distributions
print("=" * 80)
print("VARIABLE DISTRIBUTIONS")
print("=" * 80)

variables = ['temperature', 'humidity', 'precipitation', 'pressure', 
             'wind_speed', 'wind_direction', 'cloud_cover']

fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.flatten()

for idx, var in enumerate(variables):
    axes[idx].hist(df[var].dropna(), bins=50, color='steelblue', edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'{var.replace("_", " ").title()} Distribution', fontweight='bold')
    axes[idx].set_xlabel(var)
    axes[idx].set_ylabel('Frequency')
    axes[idx].grid(alpha=0.3)
    
    # Add statistics
    mean_val = df[var].mean()
    median_val = df[var].median()
    axes[idx].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
    axes[idx].axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.2f}')
    axes[idx].legend()

# Remove extra subplots
for idx in range(len(variables), len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.show()
```

```python
# Box Plots for Outlier Detection
print("=" * 80)
print("OUTLIER DETECTION")
print("=" * 80)

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

for idx, var in enumerate(variables):
    sns.boxplot(y=df[var], ax=axes[idx], color='lightcoral')
    axes[idx].set_title(f'{var.replace("_", " ").title()}', fontweight='bold')
    axes[idx].set_ylabel(var)
    axes[idx].grid(alpha=0.3)
    
    # Calculate outliers
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[var] < Q1 - 1.5*IQR) | (df[var] > Q3 + 1.5*IQR)][var]
    axes[idx].text(0.5, 0.95, f'Outliers: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)', 
                   transform=axes[idx].transAxes, ha='center', va='top',
                   bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Remove extra subplot
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis
print("=" * 80)
print("CORRELATION ANALYSIS")
print("=" * 80)

# Calculate correlation matrix
correlation_vars = ['temperature', 'humidity', 'precipitation', 'pressure', 
                    'wind_speed', 'cloud_cover']
corr_matrix = df[correlation_vars].corr()

print("\nCorrelation Matrix:")
display(corr_matrix)

# Visualize correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Weather Variables Correlation Matrix', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Find strong correlations
print("\nStrong Correlations (|r| > 0.5):")
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > 0.5:
            print(f"  {corr_matrix.columns[i]} vs {corr_matrix.columns[j]}: {corr_matrix.iloc[i, j]:.3f}")
```

```python
# Key Findings Summary
print("=" * 80)
print("KEY FINDINGS SUMMARY")
print("=" * 80)

findings = f"""
üìä DATA EXPLORATION SUMMARY

1. DATASET OVERVIEW:
   - Total Records: {len(df):,}
   - Locations: {df['location_name'].nunique()}
   - Date Range: {df['datetime'].min()} to {df['datetime'].max()}
   - Duration: {(df['datetime'].max() - df['datetime'].min()).days} days

2. DATA QUALITY:
   - Missing Values: {df.isnull().sum().sum()} ({df.isnull().sum().sum()/df.size*100:.2f}%)
   - Duplicate Records: {df.duplicated().sum()}
   - Data Completeness: {(1 - df.isnull().sum().sum()/df.size)*100:.2f}%

3. TEMPERATURE:
   - Mean: {df['temperature'].mean():.2f}¬∞C
   - Range: {df['temperature'].min():.2f}¬∞C to {df['temperature'].max():.2f}¬∞C
   - Std Dev: {df['temperature'].std():.2f}¬∞C

4. PRECIPITATION:
   - Total: {df['precipitation'].sum():.2f} mm
   - Average Daily: {df.groupby('date')['precipitation'].sum().mean():.2f} mm
   - Days with Rain: {(df.groupby('date')['precipitation'].sum() > 0).sum()}

5. STRONGEST CORRELATIONS:
"""

# Add top correlations
corr_matrix_flat = corr_matrix.unstack()
corr_matrix_flat = corr_matrix_flat[corr_matrix_flat != 1]
top_corr = corr_matrix_flat.abs().sort_values(ascending=False).head(3)

for idx, (pair, value) in enumerate(top_corr.items(), 1):
    findings += f"   {idx}. {pair[0]} vs {pair[1]}: {corr_matrix.loc[pair[0], pair[1]]:.3f}\n"

print(findings)

# Save findings to file
os.makedirs('../data/processed', exist_ok=True)
with open('../data/processed/exploration_findings.txt', 'w', encoding='utf-8') as f:
    f.write(findings)

print("\n‚úÖ Findings saved to: data/processed/exploration_findings.txt")
```
